```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Fast = np.genfromtxt('pubmed_Fast.csv',delimiter=',')
AS = np.genfromtxt('pubmed_AS.csv',delimiter=',')
```

```python
plt.plot(AS[0,:],label = 'AS-GCN')
plt.plot(Fast[0,:],label = 'FastGCN')
#plt.plot(Vanilla_cora_train_acc,label = 'Vanilla-GCN')
plt.legend()
plt.title('Train Accuracy of 2 models over Pubmed Dataset')
plt.xlabel('Epoch [/] \n each number stand for 10 epochs (e.g. 0 stands for average of 0-9 epochs)')
plt.ylabel('Test Accuracy [/]')
```

```python
plt.plot(AS[1,:],label = 'AS-GCN')
plt.plot(Fast[1,:],label = 'FASTGCN')
#plt.plot(Vanilla_cora_test_acc,label = 'Vanilla-GCN')
plt.legend()
plt.title('Test Accuracy of 2 models over Pubmed Dataset')
plt.xlabel('Epoch [/] \n each number stand for 10 epochs (e.g. 0 stands for average of 0-9 epochs)')
plt.ylabel('Test Accuracy [/]')
```

```python
plt.plot(AS[2,:],label = 'AS-GCN')
plt.plot(Fast[2,:],label = 'FASTGCN')
#plt.plot(Vanilla_cora_train_times,label = 'Vanilla-GCN')
plt.legend()
plt.title('Train Times of 2 models over Pubmed Dataset')
plt.xlabel('Epoch [/] \n each number stand for 10 epochs (e.g. 0 stands for average of 0-9 epochs)')
plt.ylabel('Train Times [s]')
```
